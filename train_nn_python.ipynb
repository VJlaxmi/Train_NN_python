```python
## Importing all the required lib

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
```

```python
plt.close('all')
##closing all the plots if anything is present
```

```python
# create a function data
x = np.arange(0,5,0.01)
x = x.reshape(-1,1)
x
```

```python
y = x**3;
y = y.reshape(-1,1)
y
```

```python
plt.plot(x,y)
```

```python
# Create MLP regressor
net = MLPRegressor(hidden_layer_sizes=(100,), activation='logistic')
net.solver = 'lbfgs'
net.max_iter=10000
```

```python
# Traing network
net.fit(x,y)
```

```python
# predict network output
y_predict = net.predict(x)
```

```python
# plot predicted values
plt.plot(x,y_predict)
plt.legend(['original','predicted'])
```

```python
# calculate the prediction error
mape = np.sum(abs(y[:,0] - y_predict))/np.size(y)
print('Prediction error is: \n')
print(mape)
```
